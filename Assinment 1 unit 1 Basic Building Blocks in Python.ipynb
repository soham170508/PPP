Q1.Develop a Python program to accept student details such as name and marks of subjects. Calculate total marks, average, and grade using conditional statements. Use proper indentation and comments.

```python
name = input("Enter student name: ")

m1 = float(input("Enter marks of Subject 1: "))
m2 = float(input("Enter marks of Subject 2: "))
m3 = float(input("Enter marks of Subject 3: "))


total = m1 + m2 + m3
average = total / 3


if average >= 75:
    grade = "A"
elif average >= 60:
    grade = "B"
elif average >= 50:
    grade = "C"
elif average >= 40:
    grade = "D"
else:
    grade = "Fail"


print("\n--- Student Result ---")
print("Name:", name)
print("Total Marks:", total)
print("Average:", average)
print("Grade:", grade)
```

Q3.Develop a menu-driven Python program to calculate total cost of electronic tools using arithmetic operators and conditional statements.

```python
total_cost = 0

while True:
    print("\n--- Electronics Tools Menu ---")
    print("1. Multimeter - Rs. 1200")
    print("2. Oscilloscope - Rs. 25000")
    print("3. Soldering Iron - Rs. 800")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        total_cost += 1200
    elif choice == 2:
        total_cost += 25000
    elif choice == 3:
        total_cost += 800
    elif choice == 4:
        break
    else:
        print("Invalid choice")

print("\nTotal Cost of Tools: Rs.", total_cost)
```

Q4.Create a Python program that accepts document name and document type and classifies the document category using if-else or else-if ladder.

```python
doc_name = input("Enter Document Name: ")
doc_type = input("Enter Document Type (pdf/doc/xls/jpg): ")

if doc_type == "pdf":
    category = "Official Document"
elif doc_type == "doc":
    category = "Text Document"
elif doc_type == "xls":
    category = "Spreadsheet Document"
elif doc_type == "jpg":
    category = "Image Document"
else:
    category = "Unknown Document Type"

print("\n--- Document Details ---")
print("Document Name:", doc_name)
print("Document Category:", category)
```

Q5. Develop a menu-driven calculator that performs arithmetic operations using loops and decision-making statements.

```python
while True:
    print("\n--- Simple Calculator ---")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 5:
        print("Calculator Closed")
        break

    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))

    if choice == 1:
        print("Result:", num1 + num2)
    elif choice == 2:
        print("Result:", num1 - num2)
    elif choice == 3:
        print("Result:", num1 * num2)
    elif choice == 4:
        if num2 != 0:
            print("Result:", num1 / num2)
        else:
            print("Error: Division by zero")
    else:
        print("Invalid choice")
```
