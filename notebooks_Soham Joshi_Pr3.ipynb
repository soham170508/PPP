3. Count the number of digits in a given number

```python
num = int(input("Enter a number: "))
count = 0

while num != 0:
    num = num // 10
    count += 1

print("Number of digits:", count)
```

2. Find the sum of digits of a number

```python
num = int(input("Enter a number: "))
sum_digits = 0

while num > 0:
    sum_digits += num % 10
    num //= 10

print("Sum of digits:", sum_digits)
```

3. Reverse a given number

```python
num = int(input("Enter a number: "))
rev = 0

while num > 0:
    rev = rev * 10 + num % 10
    num //= 10

print("Reversed number:", rev)
```

4. Check whether a number is a palindrome using a loop


```python
num = int(input("Enter a number: "))
temp = num
rev = 0

while temp > 0:
    rev = rev * 10 + temp % 10
    temp //= 10

if num == rev:
    print("Palindrome Number")
else:
    print("Not a Palindrome")
```

2. Create a menu-driven calculator (Addition, Subtraction, Multiplication, Division)

```python
while True:
    print("\n1.Add\n2.Subtract\n3.Multiply\n4.Divide\n5.Exit")
    choice = int(input("Enter choice: "))

    if choice == 5:
        break

    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))

    if choice == 1:
        print("Result:", a + b)
    elif choice == 2:
        print("Result:", a - b)
    elif choice == 3:
        print("Result:", a * b)
    elif choice == 4:
        print("Result:", a / b)
    else:
        print("Invalid choice")
```

2. Find the factorial of a given number using a `for` loop


```python
num = int(input("Enter a number: "))
fact = 1

for i in range(1, num + 1):
    fact = fact * i

print("Factorial =", fact)
```

3. Print the Fibonacci series up to `n` terms

```python
n = int(input("Enter number of terms: "))

a, b = 0, 1

print("Fibonacci Series:")
for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

4. Check whether a given number is prime


```python
num = int(input("Enter a number: "))

if num <= 1:
    print("Not a Prime Number")
else:
    for i in range(2, num):
        if num % i == 0:
            print("Not a Prime Number")
            break
    else:
        print("Prime Number")
```

5. Count the number of vowels in a given string using a loop

```python
text = input("Enter a string: ")
count = 0

for ch in text:
    if ch.lower() in 'aeiou':
        count += 1

print("Number of vowels:", count)
```
