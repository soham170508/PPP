Program to display RF bands using if–elif

```python
f = float(input("Enter frequency in Hz: "))

if 3000 <= f < 30000:
    print("VLF Band ")
elif 30000 <= f < 300000:
    print("LF Band ")
elif 3000000 <= f < 30000000:
    print("HF Band ")
elif 30000000 <= f < 300000000:
    print("VHF Band ")
else:
    print("Frequency not in standard RF Range")
```

```python
nominal_resistence = 100
tolerence_percent = 5

mesured_resistence = 104

tolerence = nominal_resistence * (tolerence_percent/100)
lower_limit = nominal_resistence - tolerence
upper_limit = nominal_resistence + tolerence

if lower_limit <= mesured_resistence <= upper_limit:
    print("The resistor is within tolerence.")
else:
    print("The resistor is out tolerence.")
    
```
